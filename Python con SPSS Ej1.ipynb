## 3. Paso 3:
- **Crear un DataFrame en pandas**

Creamos un `DataFrame` en pandas para manejar la información.

```python
# Creamos un DataFrame a partir de la lista de datos
df = pd.DataFrame(genero, columns=['Sexo'])
```

## 4. Paso 4:
- **Calcular la tabla de frecuencias**

Usamos la función value_counts para contar cuántos "Hombres" y cuántas "Mujeres" hay en la serie de datos.

```python
# Creamos la tabla de frecuencias
frecuencia = df['Sexo'].value_counts()
print(frecuencia)
```

## 5. Paso 5:
**Generar el gráfico de torta**

Utilizamos `matplotlib` para generar un gráfico de torta, que visualiza la proporción de hombres y mujeres.

```python
# Generamos el gráfico de torta
plt.figure(figsize=(3,3))  # Definimos el tamaño del gráfico
plt.pie(frecuencia, labels=frecuencia.index, autopct='%1.1f%%', startangle=90, colors=['#66b3ff','#ff9999'])
plt.title('Distribución de Hombres y Mujeres')
plt.axis('equal')  # Para que el gráfico de torta se vea circular
plt.show()
```
